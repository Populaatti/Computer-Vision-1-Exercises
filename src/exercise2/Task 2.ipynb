```python
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.util import random_noise
from skimage.filters import gaussian, sobel
import matplotlib.patches as patches
from PIL import Image

photo_path = './woman.png'

noisy_photo = random_noise(imread(fname=photo_path),)
gaussian_photo = gaussian(noisy_photo, sigma=1)
noisy_sobel = sobel(noisy_photo)
smoothed_sobel = sobel(gaussian_photo)

```

```python
plt.imshow(noisy_sobel)
```

```python
intensity, edges = np.histogram(noisy_sobel, bins=1000, range=(0, 1))
plt.plot(edges[0:-1], intensity)
plt.show()
```

```python
plt.imshow(smoothed_sobel)
```

```python
intensity, edges = np.histogram(smoothed_sobel, bins=1000, range=(0, 1))
plt.plot(edges[0:-1], intensity)
plt.show()
```

```python
binary_mask_smoothed = smoothed_sobel >= 0.06
plt.imshow(binary_mask_smoothed)

```

```python
binary_mask_noisy = noisy_sobel > 0.15
plt.imshow(binary_mask_noisy)
```

Smoothing helps with edge detection because it smooths image by removing low contrast difference points like noise. When edge detection is later  applied to the image, clearer contrast differences like edges don't mix up with noise.
