```python
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.feature import match_template
from skimage.color import rgb2gray
import matplotlib.patches as patches

panorama_path = './panorama.jpg'
template_path = './template.jpg'


grayscale_panorama = rgb2gray(imread(fname=panorama_path))
grayscale_template = rgb2gray(imread(fname=template_path))
correlation = match_template(grayscale_panorama, grayscale_template)

```

```python
# Normal test
window_corner = np.unravel_index(np.argmax(correlation), np.shape(correlation))
```

```python

rect = patches.Rectangle((window_corner[::-1]), 1920, 1080, linewidth=1, edgecolor='r', facecolor='none')
      
fig, ax = plt.subplots()
ax.imshow(imread(fname=panorama_path))
ax.add_patch(rect)
plt.show()
```

```python
# Flipped test
flipped_gs_template = grayscale_template[::-1]
flipped_correlation = match_template(grayscale_panorama, flipped_gs_template)
window_corner = np.unravel_index(np.argmax(flipped_correlation), np.shape(flipped_correlation))

rect = patches.Rectangle((window_corner[::-1]), 1920, 1080, linewidth=1, edgecolor='r', facecolor='none')
      
fig, ax = plt.subplots()
ax.imshow(imread(fname=panorama_path))
ax.add_patch(rect)
plt.show()
```
